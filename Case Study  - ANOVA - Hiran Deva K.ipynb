# 3.Check whether all the zones generate the same amount of sales

null_hypothesis : mean sales of all the zones are equal

alternate_hypothesis : mean sale of atleast one zone differs from the rest

```python
F,p=stat.f_oneway(data['Zone - A'], data['Zone - B'], data['Zone - C'], data['Zone - D'])
print('F-value : ',F)
print('p-value : ',p)
```

```python
if p<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

#### Inference:
mean sale of atleast one zone differs from the rest
